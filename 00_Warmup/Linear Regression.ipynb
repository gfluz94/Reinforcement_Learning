```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.arange(50,100, 1)*np.random.uniform(low=0.9, size=50)
y = np.arange(50,100, 1)*np.random.uniform(low=0.9, size=50)
```

```python
plt.scatter(X, y);
```

```python
X = np.concatenate([np.ones(shape=(X.shape[0], 1)), X.reshape(-1,1)], axis=1)
y = y.reshape(-1,1)
```

```python
def gradient(X, w, y):
    return (2/X.shape[0])*X.T@(X@w-y)
```

```python
i = 1
learning_rate = 1e-4
max_iterations = 1000
precision = 1e-5
w = np.random.normal(size=(2, 1))

while np.linalg.norm(gradient(X, w, y), ord=2)>precision:
    print(f"Iteration #{i}")
    last_grad = np.linalg.norm(gradient(X, w, y), ord=2)
    w -= learning_rate*gradient(X, w, y)
    loss = np.mean((y-X@w)**2)
    print(f"\tWeights: {w[:,0]}")
    print(f"\tMSE: {loss:.4f}")
    print(f"\tGradient 2-Norm: {np.linalg.norm(gradient(X, w, y), ord=2):.4f}")
    if i==max_iterations or abs(last_grad-np.linalg.norm(gradient(X, w, y), ord=2))<precision:
        break
    i+=1
```

```python
plt.scatter(X[:,1], y)
plt.plot(X[:,1], X@w, c="r");
```
